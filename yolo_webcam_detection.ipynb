```python
import torch
import cv2

import numpy as np
```

```python
modelo = torch.hub.load('ultralytics/yolov5', 'yolov5n', pretrained=True)
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    _, frame = cap.read()
    
    saida = modelo(frame)
    cv2.imshow('YOLO', np.squeeze(saida.render()))

    if cv2.waitKey(1) & 0xFF == ord('c'):
        break

cap.release()
cv2.destroyAllWindows()
```
